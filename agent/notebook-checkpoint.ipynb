```python
import pdfplumber

def text_from_pdf(pdf_file) -> str:
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
            # Parcourt chaque page
            for page_number, page in enumerate(pdf.pages, start=1):
                # Essaie d'extraire le texte “natif”
                page_text = page.extract_text() or ""
                #text += f"\n--- Page {page_number} ---\n"
                text += page_text
    return text
```

```python
import re

# 8/1/2024, 01-08-24, 2024-08-01, 2024.08.01, etc.
DATE_NUM_INLINE = re.compile(
    r'(?<!\d)(?:\d{1,2}[./-]\d{1,2}[./-]\d{2,4}|(?:19|20)\d{2}[./-]\d{1,2}[./-]\d{1,2})(?!\d)'
)

# Mois FR/EN: "12 mars 2024", "March 12, 2024"
MONTHS = r'(?:janv\.?|févr\.?|fevr\.?|mars|avr\.?|mai|juin|juil\.?|août|aout|sept\.?|oct\.?|nov\.?|d[ée]c\.?|january|february|march|april|may|june|july|august|september|october|november|december)'
DATE_TXT_INLINE = re.compile(
    rf'(?i)\b(?:\d{{1,2}}\s+{MONTHS}\s+\d{{2,4}}|{MONTHS}\s+\d{{1,2}},\s+\d{{4}})\b'
)

def strip_dates_only(text: str) -> str:
    # retire uniquement les dates, laisse tout le reste tel quel
    text = DATE_NUM_INLINE.sub('', text)
    text = DATE_TXT_INLINE.sub('', text)
    return text
```
